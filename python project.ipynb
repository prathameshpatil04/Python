```python
print("Press 1 for list operation.")
print("Press 2 for set operation.")
print("Press 3 for dictionary operation.")
print("Press 4 for tuple operation.")
print("Press 5 for string operation.")
print("\n")
ch1 = int(input("Enter the choice  for select the datatype : "))
print("****************************************************************************************")
if ch1 == 1: 
    l = list(eval(input("Create a list : ")))
    print("**************************************************************************************")
    print("Press 1 for Addition.")
    print("Press 2 for deletion .")
    print("Press 3 to get the information about list.")
    print("\n")
    ch2 = int(input("Enter the choice for list operations : "))
    print("***************************************************************************************")
    if ch2 == 1:
        print("Press 1 for 'append' method.")
        print("Press 2 for 'insert' method.")
        print("Press 3 for 'extend' method.")
        print("\n")
        ch3 = int(input("Enter the choice to select method : "))
        print("***********************************************************************************")
        if ch3 == 1:
            a = eval(input("Enter a element you want to add into list : "))
            l.append(a)
            print("List after performing append operation :",l)
        elif ch3 == 2:
            a = eval(input("Enter a element you want to add into list : "))
            b = int(input("Enter a location-index where new element were added : "))
            l.insert(b,a)
            print("***************************************************************************")
            print("List after performing insert operation :",l)
        elif ch3 == 3:
            print("******************************************************************************")
            a = list(eval(input("Enter list of multiple element to extend the orignal list : ")))
            l.extend(a)
            print("*****************************************************************************")
            print("List after performing extend operation :",l)
        else:
            print("You enterd wrong choice number.")
    elif ch2 == 2:
        print("Press 1 for remove method.")
        print("Press 2 for pop method.")
        print("Press 3 for clear method.")
        print("press 4 for delete method")
        print("\n")
        ch3 = int(input("Enter the choice to select method."))
        print("********************************************************************************")
        if ch3 == 1:
            a = eval(input("Enter a element you want to remove from the list : "))
            l.remove(a)
            print("*********************************************************************************")
            print("List after performing remove operation :",l)
        elif ch3 == 2:
            a = int(input("Enter the index of the value which want to remove from list : "))
            l.pop(a)
            print("**********************************************************************************")
            print("List after performing pop operation :",l)
        elif ch3 == 3:
            l.clear()
            print("List after performing clear operation :",l," i.e. empty list")
        elif ch3 == 4:
            del l
            print("list is deleted.")
        else:
            print("You enterd wrong choice number.")
    elif ch2 == 3:
        print(" List is a unorderd, heterogenous and mutable collection of a values. \n List are enclosed in square bracket [] and list element are seperated by comma (,)")
    else:
        print("You enterd wrong choice number.")
elif ch1 == 2:
    s = set(eval(input("Create the set : ")))
    print("***********************************************************************************")
    print("Press 1 for Addition.")
    print("Press 2 for deletion.")
    print("Press 3 to get the information about set.")
    print("\n")
    ch2 = int(input("Enter the choice for set operation : "))
    print("**********************************************************************************")
    if ch2 == 1:
        print("Press 1 for 'add' method.")
        print("Press 2 for 'update' method.")
        print("\n")
        ch3 = int(input("Enter the choice to select method : "))
        print("*****************************************************************************")
        if ch3 == 1:
            a = eval(input("Enter the value which you want to add in set : "))
            s.add(a)
            print("*************************************************************************")
            print("set after performing add operation :",s)
        elif ch3 == 2:
            a = list(eval(input("Enter multiple value in list format to add into set : ")))
            s.update(a)
            print("**************************************************************************")
            print("Set after performing update operation :",s)
        else:
            print("You enterd wrong choice number.")
    elif ch2 == 2:
        print("Press 1 for pop method.")
        print("Press 2 for discard method.")
        print("Press 3 for remove method.")
        print("Press 4 for delete method.")
        print("\n")
        ch3 = int(input("Enter the choice to select the method : "))
        print("*************************************************************************************")
        if ch3 == 1:
            s.pop()
            print("Set after performing pop operation :",s)
        elif ch3 == 2:
            a = eval(input("Enter the value remove from the set : "))
            s.discard(a)
            print("************************************************************************************")
            print("Set after performing discard operation :",s)
        elif ch3 == 3:
            a = eval(input("Enter the value remove from the set : "))
            s.remove(a)
            print("************************************************************************************")
            print("Set after performing remove operation :",s)
        elif ch3 == 4:
            del s
            print("Set is deleted.")
        else:
            print("You enterd wrong choice number.")
    elif ch2 == 3:
        print(" Set is a unordered, unique, heterogenous and mutable collection of values. \n set are enclosed in curly bracket {} and set element are seperated by comma (,)")
    else:
        print("You enterd wrong choice number.")
elif ch1 == 3:
    d = dict(eval(input("Create the dictionary : ")))
    print("***********************************************************************************************")
    print("Press 1 for Addition")
    print("Press 2 for deletion")
    print("Press 3 for access the data of dictionary.")
    print("Press 4 to get the information about dictionary.")
    print("\n")
    ch2 = int(input("Enter the choice for operation : "))
    print("**********************************************************************************************")
    if ch2 == 1:
        print("Press 1 for add values using key.")
        print("Press 2 for update method")
        print("\n")
        ch3 = int(input("Enter a choice to select the addition method : "))
        if ch3 == 1:
            a = eval(input("enter the name for key : "))
            b = eval(input("Enter thr value of key : "))
            d[a]=b
            print("Dictionary after performing operation :",d)
        elif ch3 == 2:
            a = dict(eval(input("Enter the multiple key value pair in dictionary format : ")))
            d.update(a)
            print("Dictionary after performing update operation :",d)
        else:
            print("You enterd wrong choice number.")
    elif ch2 == 2:
        print("Press 1 for the delete method.")
        print("Press 2 for the pop method.")
        print("Press 3 for the popitems method.")
        print("Press 4 for the clear method.")
        print("\n")
        ch3 = int(input("Enter the choice for the method : "))
        print("**********************************************************************************")
        if ch3 == 1:
            a = eval(input("Enter the key of record which you want to delete from dictionary : "))
            del d[a]
            print("Dictionary after a delete operation",d)
        elif ch3 == 2:
            a = eval(input("Enter the key of record which you want to delete from dictionary : "))
            d.pop(a)
            print("Dictionary after a pop operation",d)
        elif ch3 == 3:
            d.popitem()
            print("Dictionary after a popitem operation",d)
        elif ch3 == 4:
            d.clear()
            print("Dictionary after a clear operation",d," i.e. empty dictionary")
        else:
            print("You enterd wrong choice number.")
    elif ch2 == 3:
        print("Press 1 to access all the keys of dictionary.")
        print("Press 2 to access all the values of dictionary.")
        print("Press 3 to access value of perticular key based on key name.")
        print("Press 4 to access all key-value pair of dictionary.")
        print("\n")
        ch3 = int(input("Enter the choice for the method : "))
        print("********************************************************************")
        if ch3 == 1:
            a = d.keys()
            print("All keys of dictionary are :",a)
        elif ch3 == 2:
            a = d.values()
            print("All values of dictionary are :",a)
        elif ch3 == 3:
            k = eval(input("Enter a key name whose value is to be access : "))
            a = d.get(k)
            print("*********************************************************************")
            print(f"Value of the key {k} is :",a)
        elif ch3 == 4:
            a = d.items()
            print("All key-value pairs of dictionary are :",a)
        else:
            print("You enterd wrong choice number.")
    elif ch2 == 4:
        print(" Dictionary is a unordered, heterogenous and mutable group key-value pairs. \n In dictionary keys are immutable while values are mutable. \n Dictionary are enclosed in curly bracket {} and dictionary element are seperated by comma (,)")
    else:
        print("You enterd wrong choice number.")
elif ch1 == 4:
    t = tuple(eval(input("Create the tuple : ")))
    print("*************************************************************************************")
    print("Press 1 to access the single elemnt from tuple using indexing.")
    print("Press 2 to access multiple element from tuple using slicing.")
    print("Press 3 to get the information about the tuple")
    print("\n")
    ch2 = int(input("Enter the choice to select method : "))
    print("**************************************************************************************")
    if ch2 == 1:
        i = int(input("Enter the index whose value is to be access : "))
        a = t[i]
        print("***********************************************************************************")
        print(f"Value of index {i} is :",a)
    elif ch2 == 2:
        s = int(input("Enter the start index :"))
        e = int(input("Enter the end index : "))
        g = int(input("Enter the step between values : "))
        a = t[s:e:g]
        print("************************************************************************************************")
        print("Values after performing slicing according to given indexes :",a)
    elif ch2 == 3:
        print(" Tuple is a ordered, heterogenous and immutable group of elements. \n Tuple are enclosed in round bracket () and Tuple element are seperated by (,).")
    else:
        print("You enterd wrong choice number.")
elif ch1 == 5:
    st = str(input("Enter a string : "))
    print("*********************************************************************************************")
    print("Press 1 to access the single elemnt from string using indexing.")
    print("Press 2 to access multiple element from string using slicing.")
    print("Press 3 to get the information about the string.")
    print("\n")
    ch2 = int(input("Enter the choice to select method : "))
    print("*********************************************************************************")
    if ch2 == 1:
        i = int(input("Enter the index whose value is to be access : "))
        a = st[i]
        print("****************************************************************************************")
        print(f"Value of index {i} is :",a)
    elif ch2 == 2:
        s = int(input("Enter the start index :"))
        e = int(input("Enter the end index : "))
        g = int(input("Enter the step between values : "))
        a = st[s:e:g]
        print("*****************************************************************************************")
        print("Values after performing slicing according to given indexes :",a)
    elif ch2 == 3:
        print("String is a group of character which include A-Z alphabets, special chacters.")
    else:
        print("You enterd wrong choice number.")
else:
    print("You enterd wrong choice number for datatype.")
```
